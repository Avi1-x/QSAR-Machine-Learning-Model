# **Loading Dataset**

```python
import pandas as pd
b1 = pd.read_csv('Beta_Secretase_06_bioactivity_data_3class_pIC50_pubchem_fp.csv')
b1
```

```python
X = b1.drop('pIC50', axis=1)
X
```

```python
Y = b1['pIC50'].copy()
Y
```

```python
# Remove low variance features
from sklearn.feature_selection import VarianceThreshold

selection = VarianceThreshold(threshold=(0.1))    
X = selection.fit_transform(X)
X.shape
```

```python
# Data splitting
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

# **Build Regression models**

```python
from sklearn.metrics import explained_variance_score
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

**K nearest neighbors**

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(3) # Define Regressor
knn.fit(X_train, Y_train) # Train Model

# Make predictions
Y_train_pred = knn.predict(X_train)
Y_test_pred = knn.predict(X_test)

# Training set performance
knn_train_explained_variance_score = explained_variance_score(Y_train, Y_train_pred) # Calculate Explained Variance Score
knn_train_r2_score = r2_score(Y_train, Y_train_pred) # Calculate R^2
knn_train_mean_squared_error = mean_squared_error(Y_train, Y_train_pred) # Calculate MSE Score

# Test set performance
knn_test_explained_variance_score = explained_variance_score(Y_test, Y_test_pred) # Calculate Explained Variance Score
knn_test_r2_score = r2_score(Y_test, Y_test_pred) # Calculate R^2
knn_test_mean_squared_error = mean_squared_error(Y_test, Y_test_pred) # Calculate MSE Score

print('Model performance for Training set')
print('- Explained Variance Score: %s' % knn_train_explained_variance_score)
print('- R^2: %s' % knn_train_r2_score)
print('- MSE Score: %s' % knn_train_mean_squared_error)
print('----------------------------------')
print('Model performance for Test set')
print('- Explained Variance Score: %s' % knn_test_explained_variance_score)
print('- R^2: %s' % knn_test_r2_score)
print('- MSE Score: %s' % knn_test_mean_squared_error)
```

**Support vector machine (Radial basis function kernel)**

```python
from sklearn.svm import SVR

svm_rbf = SVR(gamma=2, C=1)
svm_rbf.fit(X_train, Y_train)

# Make predictions
Y_train_pred = svm_rbf.predict(X_train)
Y_test_pred = svm_rbf.predict(X_test)

# Training set performance
svm_rbf_train_explained_variance_score = explained_variance_score(Y_train, Y_train_pred) # Calculate Explained Variance Score
svm_rbf_train_r2_score = r2_score(Y_train, Y_train_pred) # Calculate R^2
svm_rbf_train_mean_squared_error = mean_squared_error(Y_train, Y_train_pred) # Calculate MSE Score

# Test set performance
svm_rbf_test_explained_variance_score = explained_variance_score(Y_test, Y_test_pred) # Calculate Explained Variance Score
svm_rbf_test_r2_score = r2_score(Y_test, Y_test_pred) # Calculate R^2
svm_rbf_test_mean_squared_error = mean_squared_error(Y_test, Y_test_pred) # Calculate MSE Score

print('Model performance for Training set')
print('- Explained Variance Score: %s' % svm_rbf_train_explained_variance_score)
print('- R^2: %s' % svm_rbf_train_r2_score)
print('- MSE Score: %s' % svm_rbf_train_mean_squared_error)
print('----------------------------------')
print('Model performance for Test set')
print('- Explained Variance Score: %s' % svm_rbf_test_explained_variance_score)
print('- R^2: %s' % svm_rbf_test_r2_score)
print('- MSE Score: %s' % svm_rbf_test_mean_squared_error)
```

**Decision tree**

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(max_depth=5) # Define Regressor
dt.fit(X_train, Y_train) # Train Model

# Make predictions
Y_train_pred = dt.predict(X_train)
Y_test_pred = dt.predict(X_test)

# Training set performance
dt_train_explained_variance_score = explained_variance_score(Y_train, Y_train_pred) # Calculate Explained Variance Score
dt_train_r2_score = r2_score(Y_train, Y_train_pred) # Calculate R^2
dt_train_mean_squared_error = mean_squared_error(Y_train, Y_train_pred) # Calculate MSE Score

# Test set performance
dt_test_explained_variance_score = explained_variance_score(Y_test, Y_test_pred) # Calculate Explained Variance Score
dt_test_r2_score = r2_score(Y_test, Y_test_pred) # Calculate R^2
dt_test_mean_squared_error = mean_squared_error(Y_test, Y_test_pred) # Calculate MSE Score

print('Model performance for Training set')
print('- Explained Variance Score: %s' % dt_train_explained_variance_score)
print('- R^2: %s' % dt_train_r2_score)
print('- MSE Score: %s' % dt_train_mean_squared_error)
print('----------------------------------')
print('Model performance for Test set')
print('- Explained Variance Score: %s' % dt_test_explained_variance_score)
print('- R^2: %s' % dt_test_r2_score)
print('- MSE Score: %s' % dt_test_mean_squared_error)
```

**Random forest**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=10) # Define Regressor
rf.fit(X_train, Y_train) # Train Model

# Make predictions
Y_train_pred = rf.predict(X_train)
Y_test_pred = rf.predict(X_test)

# Training set performance
rf_train_explained_variance_score = explained_variance_score(Y_train, Y_train_pred) # Calculate Explained Variance Score
rf_train_r2_score = r2_score(Y_train, Y_train_pred) # Calculate R^2
rf_train_mean_squared_error = mean_squared_error(Y_train, Y_train_pred) # Calculate MSE Score

# Test set performance
rf_test_explained_variance_score = explained_variance_score(Y_test, Y_test_pred) # Calculate Explained Variance Score
rf_test_r2_score = r2_score(Y_test, Y_test_pred) # Calculate R^2
rf_test_mean_squared_error = mean_squared_error(Y_test, Y_test_pred) # Calculate MSE Score

print('Model performance for Training set')
print('- Explained Variance Score: %s' % rf_train_explained_variance_score)
print('- R^2: %s' % rf_train_r2_score)
print('- MSE Score: %s' % rf_train_mean_squared_error)
print('----------------------------------')
print('Model performance for Test set')
print('- Explained Variance Score: %s' % rf_test_explained_variance_score)
print('- R^2: %s' % rf_test_r2_score)
print('- MSE Score: %s' % rf_test_mean_squared_error)
```

**Neural network**

```python
from sklearn.neural_network import MLPRegressor

mlp = MLPRegressor(alpha=1, max_iter=1000)
mlp.fit(X_train, Y_train)

# Make predictions
Y_train_pred = mlp.predict(X_train)
Y_test_pred = mlp.predict(X_test)

# Training set performance
mlp_train_explained_variance_score = explained_variance_score(Y_train, Y_train_pred) # Calculate Explained Variance Score
mlp_train_r2_score = r2_score(Y_train, Y_train_pred) # Calculate R^2
mlp_train_mean_squared_error = mean_squared_error(Y_train, Y_train_pred) # Calculate MSE Score

# Test set performance
mlp_test_explained_variance_score = explained_variance_score(Y_test, Y_test_pred) # Calculate Explained Variance Score
mlp_test_r2_score = r2_score(Y_test, Y_test_pred) # Calculate R^2
mlp_test_mean_squared_error = mean_squared_error(Y_test, Y_test_pred) # Calculate MSE Score

print('Model performance for Training set')
print('- Explained Variance Score: %s' % mlp_train_explained_variance_score)
print('- R^2: %s' % mlp_train_r2_score)
print('- MSE Score: %s' % mlp_train_mean_squared_error)
print('----------------------------------')
print('Model performance for Test set')
print('- Explained Variance Score: %s' % mlp_test_explained_variance_score)
print('- R^2: %s' % mlp_test_r2_score)
print('- MSE Score: %s' % mlp_test_mean_squared_error)
```

# **Build Stacked model**

```python
# Define estimators
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression

estimator_list = [
    ('knn',knn),
    ('svm_rbf',svm_rbf),
    ('dt',dt),
    ('rf',rf),
    ('mlp',mlp) ]

# Build stack model
stack_model = StackingRegressor(
    estimators=estimator_list, final_estimator=LinearRegression()
)

# Train stacked model
stack_model.fit(X_train, Y_train)

# Make predictions
Y_train_pred = stack_model.predict(X_train)
Y_test_pred = stack_model.predict(X_test)

# Training set model performance
stack_model_train_explained_variance_score = explained_variance_score(Y_train, Y_train_pred) # Calculate Explained Variance Score
stack_model_train_r2_score = r2_score(Y_train, Y_train_pred) # Calculate R^2
stack_model_train_mean_squared_error = mean_squared_error(Y_train, Y_train_pred) # Calculate MSE Score

# Test set model performance
stack_model_test_explained_variance_score = explained_variance_score(Y_test, Y_test_pred) # Calculate Explained Variance Score
stack_model_test_r2_score = r2_score(Y_test, Y_test_pred) # Calculate R^2
stack_model_test_mean_squared_error = mean_squared_error(Y_test, Y_test_pred) # Calculate MSE Score

print('Model performance for Training set')
print('- Explained Variance Score: %s' % stack_model_train_explained_variance_score)
print('- R^2: %s' % stack_model_train_r2_score)
print('- MSE Score: %s' % stack_model_train_mean_squared_error)
print('----------------------------------')
print('Model performance for Test set')
print('- Explained Variance Score: %s' % stack_model_test_explained_variance_score)
print('- R^2: %s' % stack_model_test_r2_score)
print('- MSE Score: %s' % stack_model_test_mean_squared_error)
```

# **Results**

```python
evs_train_list = {'knn':knn_train_explained_variance_score,
'svm_rbf': svm_rbf_train_explained_variance_score,
'dt': dt_train_explained_variance_score,
'rf': rf_train_explained_variance_score,
'mlp': mlp_train_explained_variance_score,
'stack': stack_model_train_explained_variance_score}

r2_train_list = {'knn':knn_train_r2_score,
'svm_rbf': svm_rbf_train_r2_score,
'dt': dt_train_r2_score,
'rf': rf_train_r2_score,
'mlp': mlp_train_r2_score,
'stack': stack_model_train_r2_score}

mse_train_list = {'knn':knn_train_mean_squared_error,
'svm_rbf': svm_rbf_train_mean_squared_error,
'dt': dt_train_mean_squared_error,
'rf': rf_train_mean_squared_error,
'mlp': mlp_train_mean_squared_error,
'stack': stack_model_train_mean_squared_error}
```
